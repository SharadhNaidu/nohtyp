```python
#Lists in python
name = ["Sharadh" , "Daredevil" , "Bullseye" , "Baymax" , "Fisk"]
#slicing is similiar to that in strings 
#lists in python are mutable in the sense they can be edited

#2D lists
matrix = [[1 , 2 , 3],[4 , 5 , 6],[7 , 8 , 9]]
print(matrix[0][0])

#Methods for lists
numbers = [1 , 2, 3 , 4, 5, 6]
#add a number to end of a list
numbers.append(7)
print(numbers)

#add a number at specific index of a list
numbers.insert(0,10)
print(numbers)

#remove a number from the list
numbers.remove(10)
print(numbers)

#get the index of a certain object
print(numbers.index(7))

#remove just the last element from, the list 
numbers.pop()
print(numbers)

# sort a list in python
numbers.reverse()
print(numbers)
numbers.sort()
print(numbers)

#get a copy of a list
number2 = numbers.copy()
print(number2)
```

```python
#Tuoples in Python 

#Tuples are like lists but tuples are immutable

#declaring a tuple

numbers = (1 , 2  , 3 , 4)

#index in tuple
print(numbers.index(2))

#count in tuple
print(numbers.count(3))
```

```python
#unpacking in python
l = [1,2,3,4,5]
x = l[0]
y = l[1]
z = l[2]
a = l[3]
b = l[4]

#instead of doing such long tedious assigning of variables froom python datastructures we use unpacking in python

x , y , z , a , b = l

print(l)
print(x , y , z , a , b)
```
